```python
import numpy as np

```

```python
# Crear un tensor de 5 dimensiones con forma (2, 3, 4, 5, 6)
tensor_5d = np.random.rand(1, 2, 3, 4, 5)

print(tensor_5d)
print("Forma del tensor:", tensor_5d.shape)
```

```python
# Añadir una nueva dimensión (de tamaño 1) al tensor
tensor_6d = np.expand_dims(tensor_5d, axis=0)

print("Forma del tensor original:", tensor_5d.shape)
print("Forma del tensor después de añadir una dimensión:", tensor_6d.shape)
```

```python
print(tensor_6d.shape) #imprime la forma del nuevo tensor
print(tensor_6d) #muestra el contenido del tensor
```

```python
# Eliminar las dimensiones de tamaño 1 (en este tensor hay dos "1")
tensor_4d = np.squeeze(tensor_6d)

# Verificar la forma del tensor después de eliminar la dimensión
print("Forma del tensor después de eliminar la dimensión:", tensor_4d.shape)
```
